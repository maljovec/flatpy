```python
import flatpy
import altair as alt
import numpy as np
import pandas as pd
alt.renderers.enable('notebook')
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Test 2D Functions

```python
%matplotlib inline

X = flatpy.utils.generate_test_grid_2d(40)
x,y = flatpy.utils.unpack2D(X)

for name, foo in flatpy.twoD.__dict__.items():
    if name in ["unpack2D"]:
        continue
    if not name.startswith("__") and type(foo).__name__ == "function" and not name.startswith("local"):
        z = foo(X)
        plt.figure()
        img = plt.tricontourf(x,y,z, cmap="cividis")
        plt.colorbar(img)
        plt.gca().set_title(name)
        plt.savefig("functions/" + name + ".png")
```

# Test nD Functions in 2D

```python
%matplotlib inline

X = flatpy.utils.generate_test_grid_2d(40)
x,y = flatpy.utils.unpack2D(X)

for name, foo in flatpy.nD.__dict__.items():
    if not name.startswith("__") and type(foo).__name__ == "function":
        z = foo(X)
        plt.figure()
        img = plt.tricontourf(x,y,z, cmap="cividis")
        plt.colorbar(img)
        plt.gca().set_title(name)
        plt.savefig("functions/" + name + ".png")
```

# 3D Renderings

```python
%matplotlib notebook

X = flatpy.utils.generate_test_grid_2d(40)
x,y = flatpy.utils.unpack2D(X)

for name, foo in flatpy.twoD.__dict__.items():
    if name in ["unpack2D"]:
        continue
    if not name.startswith("__") and type(foo).__name__ == "function" and not name.startswith("local"):
        z = foo(X)
        fig = plt.figure()
        ax = fig.gca(projection='3d')
        img = ax.plot_trisurf(x, y, z, cmap="cividis")
        plt.colorbar(img)
        plt.gca().set_title(name)
        plt.savefig("functions/" + name + "_3D.png")

for name, foo in flatpy.nD.__dict__.items():
    if not name.startswith("__") and type(foo).__name__ == "function":
        z = foo(X)
        fig = plt.figure()
        ax = fig.gca(projection='3d')
        img = ax.plot_trisurf(x, y, z, cmap="cividis")
        plt.colorbar(img)
        plt.gca().set_title(name)
        plt.savefig("functions/" + name + "_3D.png")
```
